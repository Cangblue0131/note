## ANOVA (變異數分析)
<font size = 5 color = red>目的 : 檢定其各類別間平均數是否相等</font>

<font size = 4 color = orange >迴歸分析的 ANOVA 表和此不同，參考 "監督式學習-> Regression-> Multiple regression"。</font>

[參考網址:wiki](https://zh.wikipedia.org/zh-tw/%E6%96%B9%E5%B7%AE%E5%88%86%E6%9E%90#ANOVA%E7%9A%84%E6%A8%A1%E5%BC%8F%E5%81%87%E8%A8%AD)

參考書籍 : 統計學 - 李德治、童惠玲

### 前提假設
*   自變數(又稱解釋變數、獨立變數)為類別變數(categorical variable)
*   依變數(又稱應變數、因變數)必須是連續變數(continuous variable)
*   <font color = red>母群體必須是常態分佈(Normal Distribution)</font>
*   獨立事件(Independent event)：樣本須為獨立變項(Independent variable)
*   變異數(Variance)同質性：<font color = red>樣本間的變異數必須相等。</font>

### 單因子變異數分析(One-way ANOVA)：只有一個自變項的變異數分析。

因子指的是分類性，所以單因子代表只有一項分類（例如班級、血型等）。

模型可寫成 : $y_{ij} = \mu_{i} + \varepsilon_{ij}$\
其中 : $y_{ij} \sim N(\mu_i,\sigma^2) ~;~ \varepsilon_{ij} \sim N(0,\sigma^2)$ 且 $\varepsilon_{ij}$ 互為獨立 , $i = 1,2,...k ~;~ j = 1,2,...,n$

$\alpha_i = \mu_i - \mu$ : 第i個處理效應(treatment effect)，並且$\sum\alpha = 0$。\
所以模型可以改寫成 : $y_{ij} = \mu + \alpha_i + \varepsilon_{ij}$

&emsp;

$H_0 : 各組母體平均數都相等$\
$H_\alpha : 並非各組母體平均數都相等$

資料格式:\
注: 每個母體不一定是 $n$ 個樣本，所以通常會表示成 $n_1,n_2,...,n_k$

<img src="https://i.imgur.com/AbegM2I.png" width="40%" height="40%">


#### One-way ANOVA table

<font size = 5>符號 : </font>

*   k : 組數

*   N : 總樣本數 ( $\sum n_i$ )

*   $y_{k,i}$ : 第k組第i個觀察值 (上面表示成 $x_{i,k}$)

*   $\bar{y_{k}}$ : 第k組觀察值的平均值 (上面表示成 $\bar{x_{k}}$)

*   $\bar{y}$    : 所有觀察值的平均值 (上面表示成 $\bar{\bar{x}}$)

*   $SS_T (總變異) : \sum\limits_{k}\sum\limits_{i}(y_{k,i} - \bar{y})^2 ~=~ n_T\sigma_T^2 ~=~ (n_T-1)s_T^2 $

*   $SS_G (組間變異, SS_A (A 因子的變異)) : \sum\limits_{k}n_k(\bar{y_{k}} - \bar{y})^2 = SS_T - SS_W = SS_T - SS_E$

*   $SS_W (組內變異, SS_E) : \sum\limits_{k}\sum\limits_{i}(y_{k,i} - \bar{y_{k}})^2  = \sum\limits_{k}(n_k-1)(\frac{1}{n_k-1}\sum\limits_{i}(y_{k,i} - \bar{y_{k}})^2) = \sum\limits_{k} (n_k-1)s_k^2$

*   $SS_T = SS_G + SS_W$

*   $F^*$   : 若 $F^* > F_\alpha(k-1,n-k)$，則 $ reject ~~H_0 $ 。


<img src="https://i.imgur.com/R2OoUE9.png" width="60%" height="60%">

#### python

```python
from pickle import FALSE
import numpy as np
import pandas as pd

##input
G = 12 #組別



##建立資料
data = {}
n = []
for i in range(G):
    ni = np.random.randint(12,30)
    data["母體{}".format(i+1)] = np.random.normal(loc = 0, scale = 1 ,size = ni)
    n.append(ni)
N = np.sum(n) #總樣本數

##處理資料(python 不同長度製作成data frame很麻煩，所以直接處理資料在製作，之後計算也比較方便)
pre_data = {"ni"   : [ len(data[i]) for i in data.keys()],
            'xbar' : [ np.mean(data[i]) for i in data.keys()] ,
            's2'  :  [ np.var(data[i],ddof = 1) for i in data.keys()]}
df = pd.DataFrame(pre_data)
df.index = data.keys()
df

##run
all_mean = df['ni'] * df['xbar']/np.sum(df['ni']) #sum(ni*xbar/sum(ni))

SSG =np.sum(df['ni']*(df['xbar']-all_mean)**2)  #sum(nk(xbar-xbarbar))
SSW = np.sum((df['ni']-1) * df['s2'] ) #sum((nk-1)*sk^2)
SST = SSG + SSW

SS = [SSG,SSW,SST]
df = [G-1,N-G,N-1]
MS = [SS[0]/df[0],SS[1]/df[1],np.NaN]
F = [ MS[0]/MS[1],np.NaN,np.NAN]

ANOVA_df = pd.DataFrame([SS,df,MS,F]).T
ANOVA_df.index = ["因子(組間)","隨機(組內)","總和"]
ANOVA_df.columns = ["平方和","自由度","平均平方和","F"]

ANOVA_df #用print 會沒對齊

```

### 雙因子變異數分析 (two way ANOVA)

雙因子代表有兩種分類性(A因子、B因子)，例如血型 (A,B,AB,O) 和 性別，所以種個有 4*2 = 8 種類別。

*   簡述:
    
    two way ANOVA 分成 <font size = 4 color = red>未重複實驗</font> 和 <font size = 4 color = red>重複實驗</font>，差別在於每個種類是否僅抽取一個觀察值 (未重複試驗 例如 :A1肥料B1品種只取1次收穫量,A2B1也是...)。


#### 未重複實驗

*   假設檢定 : 分為 A 和 B 因子的假設
    *   A因子的假設
    
        $H_0 : \mu_{A_1} = \mu_{A_2} = ... = \mu_{A_c}\\$
        $H_0 : \mu_{A_1} , \mu_{A_2} , ... , \mu_{A_c}$ 不全相等
    
    *   B因子的假設
    
        $H_0 : \mu_{B_1} = \mu_{B_2} = ... = \mu_{B_r}\\$
        $H_0 : \mu_{B_1} , \mu_{B_2} , ... , \mu_{B_r}$ 不全相等
    


*   資料格式:

<img src="https://i.imgur.com/DCbvav9.png" width="60%" height="60%">


##### table

因為有兩個因子，所以總變異可以分解成：總變異 = A因子引起的變異(SSA) + B因子引起的變異(SSB) + 隨機變異(SSE)

*   $c : $ A 因子種類

*   $r : $ B 因子種類

*   $N : $ 總樣本數

*   $SS_T (總變異) : \sum\limits_{i=1}^r\sum\limits_{j=1}^c(x_{i,j} - \bar{\bar{x}})^2 ~=~ 
                            \sum\limits_{i=1}^r\sum\limits_{j=1}^c x_{i,j}^2 -n_T \bar{\bar{x}}^2 ~=~ n_T\sigma_T^2   = SSA + SSB + SSE$

*   $SSA : \sum\limits_{i=1}^r\sum\limits_{j=1}^c(\bar{A}_{j} - \bar{\bar{x}})^2 ~=~ r\sum\limits_{j=1}^c\bar{A}_{j} - n_T \bar{\bar{x}}^2 ~=~ n_T\sigma_{A_i}^2$

*   $SSB : \sum\limits_{i=1}^r\sum\limits_{j=1}^c(\bar{B}_{i} - \bar{\bar{x}})^2 ~=~ c\sum\limits_{i=1}^r\bar{B}_{i} - n_T \bar{\bar{x}}^2 ~=~ n_T\sigma_{B_j}^2$

*   $SSE = \sum\limits_{i=1}^r\sum\limits_{j=1}^c(x_{ij} - \bar{A}_{j} - \bar{B}_{i} + \bar{\bar{x}})^2 =SST - SSA - SSB$

1.  檢定A因子是否有影響 : \
    若 $F^*_A > F_{\alpha,(c-1),(r-1)(c-1)} => reject~H_0 $
    
2.  檢定B因子是否有影響 : \
    若 $F^*_B > F_{\alpha,(r-1),(r-1)(c-1)} => reject~H_0 $

##### python

```python
from pickle import FALSE
import numpy as np
import pandas as pd

"""
###模擬資料
##input
Gi = 7 #組別1
Gj = 5 #組別2


##建立資料
N = Gi * Gj

DF = pd.DataFrame([list(np.random.normal(loc = 0, scale = 1 ,size = Gj)) for i in range(Gi)])
DF.index = ["B{}".format(i+1) for i in range(Gi)]
DF.columns = ["A{}".format(i+1) for i in range(Gj)]
df_mean_col = pd.DataFrame( [[np.mean(DF[i]) for i in DF.columns]], index= ["平均"], columns=DF.columns)
DF = DF.append(df_mean_col)
df_mean_row = pd.DataFrame( list(np.mean(DF.T)) ,index = DF.index, columns= ["平均"])
DF = pd.concat([DF,df_mean_row],axis = 1)

# DF[0:(len(DF)-1)][DF.columns[0:-1]]
"""

###例題資料
DF = pd.DataFrame([[3.7,4.5,3.1],[3.4,3.9,2.8],[3.5,4.1,3.0],[3.2,3.5,2.6],[3.9,4.8,3.4]])
DF.index = ["{}".format(i+1) for i in range(len(DF.index))]
DF.columns = ["低速","中速","高速"]
df_mean_col = pd.DataFrame( [[np.mean(DF[i]) for i in DF.columns]], index= ["平均"], columns=DF.columns)
DF = DF.append(df_mean_col)
df_mean_row = pd.DataFrame( list(round(np.mean(DF.T),2)) ,index = DF.index, columns= ["平均"])
DF = pd.concat([DF,df_mean_row],axis = 1)



class ANOVA_two_nrep:
    def __init__(self,x):
        """
        x 輸入DataFrame (不包含平均值)
        """
        self.x = x

    def SST(self):
        S1 = np.sum(self.x.values**2) #sum(sum(x^2))
        S2 = len(self.x.index)*len(self.x.columns) * np.mean(self.x.values)**2 #n*x_double_bar^2
        return round(S1 - S2,3)

    def SSA(self):
        S1 = np.sum(self.x.mean(axis = 0)**2)  * len(self.x.index)
        S2 = len(self.x.index)*len(self.x.columns) * np.mean(self.x.values)**2 #n*x_double_bar^2
        return round(S1 - S2,3)

    def SSB(self):
        S1 = np.sum(self.x.mean(axis = 1)**2)  * len(self.x.columns)
        S2 = len(self.x.index)*len(self.x.columns) * np.mean(self.x.values)**2 #n*x_double_bar^2
        return round(S1 - S2,3)
    
    def SSE(self):
        return round(self.SST() - self.SSA() - self.SSB(),3)
    
    def ANOVA_table(self):
        A_SS = self.SSA()
        B_SS = self.SSB()
        E_SS = self.SSE()
        T_SS = self.SST()

        A_DF = len(self.x.columns) -1
        B_DF = len(self.x.index) -1
        E_DF = A_DF * B_DF
        T_DF = len(self.x.index)*len(self.x.columns) - 1 

        A_MS = A_SS / A_DF
        B_MS = B_SS / B_DF
        E_MS = E_SS / E_DF

        A_F = A_MS / E_MS
        B_F = B_MS / E_MS

        return pd.DataFrame([[A_SS,A_DF,A_MS,A_F],
                             [B_SS,B_DF,B_MS,B_F],
                             [E_SS,E_DF,E_MS,np.NaN],
                             [T_SS,T_DF,np.NaN,np.NaN]])

ANOVA_ = ANOVA_two_nrep(DF[0:(len(DF)-1)][DF.columns[0:-1]])
ANOVA_table = ANOVA_.ANOVA_table()
ANOVA_table.index = ["速度","供應商","隨機","總和"]
ANOVA_table.columns = ["平方和","自由度","平均平方和","F"] 
ANOVA_table
```

#### 重複實驗

*   衡量水準 : 各個因素下的可能性

*   細格平均數(cell means) : 因素的衡量水準組合 (以下圖為例，有$r\times c$種) 所計算出的平均

*   變異來源 : 
    *   由細格平均數產生的變異造成交互作用的效果
    *   由 A(B) 因子產生的變異稱為 A(B) 因子主要效果

*   假設檢定 : 分為 A 、 B 因子之主要效果的假設和 AB 交互效果之假設
    *   A因子之的主要效果檢定
    
        $H_0 : \mu_{A_1} = \mu_{A_2} = ... = \mu_{A_c}\\$
        $H_0 : \mu_{A_1} , \mu_{A_2} , ... , \mu_{A_c}$ 不全相等
    
    *   B因子之的主要效果檢定
    
        $H_0 : \mu_{B_1} = \mu_{B_2} = ... = \mu_{B_r}\\$
        $H_0 : \mu_{B_1} , \mu_{B_2} , ... , \mu_{B_r}$ 不全相等
    
    *   AB 交互效果檢定

        $H_0 : A,B 因子不具交互影響\\$
        $H_1 : A,B 因子具交互影響$
    
*   資料格式 :


<img src="https://i.imgur.com/ZZVgxHB.png" width="60%" height="60%">

##### table

*   c : 因素 A 的處理數

*   r : 因素 B 的處理數

*   n : 重複數

*   N : 總樣本數

*   $x_{ijk}$ : 因素 (B,A) 的第 (i,j) 個水準的第 k 個重複值

*   $\bar{A}_j$ : 因素 A 第j個水準之樣本平均數 (表上就是第 j 行的總平均)

*   $\bar{B}_i$ : 因素 B 第i個水準之樣本平均數 (表上就是第 i 列的總平均)

*   $\overline{B_iA_j}$ : 因素 (B,A) 的第 (i,j) 個水準之組合的樣本平均數，即細格平均數

*   $\bar{\bar{x}}$ : 總平均

*   總變異數 : A 因子引起的變異 + B 因子引起的變異 + AB 因子交互作用引起的變異 + 隨機變異 $(SST = SSA + SSB + SSAB + SSE)$

*   $SS_T (總變異) : \sum\limits_{i=1}^r\sum\limits_{j=1}^c\sum\limits_{k=1}^n(x_{i,j,k} - \bar{\bar{x}})^2 ~=~ 
                            \sum\limits_{i=1}^r\sum\limits_{j=1}^c \sum\limits_{k=1}^nx_{i,j,k}^2 -n_T \bar{\bar{x}}^2 ~=~ n_T\sigma_T^2   = SSA + SSB + SSAB + SSE$

*   $SSA : rn\sum\limits_{j=1}^c(\bar{A}_{j} - \bar{\bar{x}})^2 ~=~ rn\sum\limits_{j=1}^c\bar{A}_{j} - n_T \bar{\bar{x}}^2 ~=~ n_T\sigma_{A_i}^2$

*   $SSB : cn\sum\limits_{i=1}^r(\bar{B}_{i} - \bar{\bar{x}})^2 ~=~ cn\sum\limits_{i=1}^r\bar{B}_{i} - n_T \bar{\bar{x}}^2 ~=~ n_T\sigma_{B_j}^2$

*   $SSE = \sum\limits_{i=1}^r\sum\limits_{j=1}^c\sum\limits_{k=1}^n(x_{ijk}-\overline{A_jB_i})^2 ~=~ 
                        \sum\limits_{i=1}^r\sum\limits_{j=1}^c\sum\limits_{k=1}^nx_{ijk}^2 - n\sum\limits_{i=1}^r\sum\limits_{j=1}^c\overline{A_jB_i}$

*   $SSAB = \sum\limits_{i=1}^r\sum\limits_{j=1}^c\sum\limits_{k=1}^n(\overline{A_jB_i} - \bar{A}_i - \bar{B}_j - \bar{\bar{x}})^2 = SST - SSA -SSB -SSE$

##### 檢定流程

*   交互效果檢定
    1.  若不拒絕 $H_0$
        *   A 因子主要效果檢定
            *   若拒絕 $H_0$ -> A 因子檢定 (單因子變異數分析)
        *   B 因子主要效果檢定
            *   若拒絕 $H_0$ -> B 因子檢定 (單因子變異數分析)
    2.  若拒絕 $H_0$
        *   循序方格檢定
        *   A 因子條件檢定
        *   B 因子條件檢定

##### python

```python
import numpy as np
import pandas as pd

###例題資料 p.650
data = np.array([[[109,110],[110,115],[108,110],[110,106]],
                 [[110,112],[110,111],[112,109],[114,112]],
                 [[116,114],[112,115],[114,119],[120,117]]])

### function
def make_DF(data):
    r = len(data)
    DF = pd.DataFrame([list(np.mean(data[i],axis = 1)) for i in range(r)]) #製作各細格平均數的dataframe

    df_mean_col = pd.DataFrame(round(DF.mean(axis = 0),3),columns = ["行平均"]).T
    DF = pd.concat([DF,df_mean_col],axis = 0)

    df_mean_row = pd.DataFrame(round(DF.mean(axis = 1),3),columns = ['列平均'])
    DF = pd.concat([DF,df_mean_row],axis = 1)
    
    return DF


def find_std(data,T_mean): #尋找std
    return np.round(np.sum((data - T_mean)**2 / np.size(data)),3)


class ANOVA_two_rep:
    def __init__(self,data,x):
        """
        x 輸入細格平均數的data frame
        """
        self.data = data
        self.x = x
        self.r = len(data)
        self.c = len(data[0])
        self.n = len(data[0,0])
        self.N = self.r * self.c * self.n

        self.std_T = find_std(data,x.iloc[-1,-1])
        self.std_A = find_std(x.iloc[-1,0:-1].values,x.iloc[-1,-1])
        self.std_B = find_std(x.iloc[0:-1,-1].values,x.iloc[-1,-1])

    ### SS
    def SST(self):
        return self.N * self.std_T
    
    def SSA(self):
        return self.N * self.std_A
    
    def SSB(self):
        return self.N * self.std_B
    
    def SSE(self):
        S1 = np.sum(self.data**2)
        S2 = self.n * (np.sum(self.x.iloc[0:-1,0:-1].values**2))
        return S1 - S2

    def SSAB(self):
        return round(self.SST() - self.SSA() - self.SSB() - self.SSE(),3)
    
    ### ANOVA table
    def ANOVA_table(self):
        ## SS
        SS_T = self.SST()
        SS_A = self.SSA()
        SS_B = self.SSB()
        SS_E = self.SSE()
        SS_AB = SS_T - SS_A - SS_B - SS_E #用 function 會再算一次SST,SSA,SSB,SSE

        ## DF
        DF_T = self.N - 1
        DF_A = self.c - 1
        DF_B = self.r - 1
        DF_AB = (self.c-1)*(self.r-1)
        DF_E = self.r * self.c * (self.n - 1)

        ## MS
        MS_A =  SS_A  / DF_A
        MS_B =  SS_B  / DF_B
        MS_AB = SS_AB / DF_AB
        MS_E =  SS_E  / DF_E

        ## F
        F_A  = MS_A  / MS_E
        F_B  = MS_B  / MS_E
        F_AB = MS_AB / MS_E
        
        return pd.DataFrame([[SS_A,DF_A,MS_A,F_A],
                             [SS_B,DF_B,MS_B,F_B],
                             [SS_AB,DF_AB,MS_AB,F_AB],
                             [SS_E,DF_E,MS_E,np.nan],
                             [SS_T,DF_T,np.NaN,np.NaN]])
        


DF = make_DF(data)
DF.index = ['1','2','3',"行平均"]
DF.columns = ['甲','乙','丙','丁','列平均']

DF
```

```python
ANOVA_ = ANOVA_two_rep(data,DF)
ANOVA_table = ANOVA_.ANOVA_table()
ANOVA_table.index = ["機器","操作員","交互作用","隨機","總變異"]
ANOVA_table.columns = ["平方和","自由度","平均平方和","F"] 
ANOVA_table
```
