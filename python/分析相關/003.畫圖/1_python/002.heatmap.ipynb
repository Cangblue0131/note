## 基本繪製 heatmap

建議使用 R 繪製.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.gridspec as gridspec

# 生成隨機數據
np.random.seed(42)
data = np.random.uniform(-3, 3, (20, 10))
y = np.random.choice([0, 1], size=20)

# 將數據轉換成DataFrame
df = pd.DataFrame(data, columns=[f'feature_{i+1}' for i in range(10)])
df['y'] = y

# 定義類別的顏色
def class_colour(y):
    if y == 0:
        return "#FF0000"  # 紅色
    else:
        return "#0000FF"  # 藍色

# 將顏色應用於每個類別
class_colours = [class_colour(y_val) for y_val in df['y']]

# 設置繪圖區域大小
fig = plt.figure(figsize=(12, 8))
gs = gridspec.GridSpec(1, 2, width_ratios=[5, 1])  # 調整子圖寬度的比例

# 左邊的子圖
ax1 = plt.subplot(gs[0])
sns.heatmap(data=df.drop(['y'], axis=1), cmap='coolwarm', annot=True, fmt='.2f', cbar=False, ax=ax1)
ax1.set_title('x')

# 右邊的子圖
ax2 = plt.subplot(gs[1])
y_matrix = y.reshape(-1, 1)
sns.heatmap(data=y_matrix, cmap='coolwarm', annot=True, fmt='d', cbar=False, ax=ax2)
ax2.set_title('y')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# 生成隨機數據
np.random.seed(42)
data = np.random.uniform(-3, 3, (20, 10))
y = np.random.choice([0, 1], size=20)

# 將數據轉換成DataFrame
df = pd.DataFrame(data, columns=[f'feature_{i+1}' for i in range(10)])
df['y'] = y

# 將 y=1 的行放在前面，y=0 的行放在後面
df_sorted = df.sort_values(by='y', ascending=False)

# 使用自定義函數繪製熱圖
plot_heatmaps(data=df_sorted, y=df_sorted['y'].values)
```
