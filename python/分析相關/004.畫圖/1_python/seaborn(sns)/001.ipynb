```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
DF = pd.read_csv('D:/dataset/House Prices/train.csv')
print(f'資料的維度為 : {np.shape(DF)}')
DF.head(5)
```

```python
quantitative = [feature for feature in DF.columns if DF.dtypes[feature] != 'object'] # 定量变量
quantitative.remove('Id')
quantitative.remove('SalePrice')
qualitative = [feature for feature in DF.columns if DF.dtypes[feature] == 'object'] # 定性变量
```

```python
import warnings
# 忽略 FutureWarning
warnings.simplefilter(action='ignore', category=FutureWarning)

# 定量变量直方图
m_cont = pd.melt(DF, value_vars=quantitative)
g = sns.FacetGrid(m_cont, col='variable', col_wrap=4, sharex=False, sharey=False)
g.map(sns.histplot, 'value')
```
